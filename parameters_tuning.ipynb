# Parameters tuning
Find the optimal parameter for each algorithm


```python
%load_ext autoreload
%autoreload 2

from evaluator import evaluate
from bandits import linucb, ucb1, egreedy
import dataset
from matplotlib import pyplot as plt
import numpy as np
```

```python
files = ("ydata-fp-td-clicks-v1_0.20090501","ydata-fp-td-clicks-v1_0.20090509",)
dataset.get_yahoo_events(files)

rnd_ctr, _ = evaluate(egreedy(1), mode="deploy")
```

```python
epsilon_values = np.arange(0.1,1,0.1)
tune_alpha = [egreedy(e) for e in epsilon_values]

ctrs = []
for test in tune_alpha:
    ctr, _ = evaluate(test, verbose=3, mode="learn")
    ctrs.append(ctr / rnd_ctr)

plt.plot(epsilon_values, ctrs)
plt.legend(["ε-greedy"])
plt.xlabel("ε")
plt.ylabel("CTR lift")
plt.title("Learning bucket")
plt.show()
```

```python
ctrs = []
for test in tune_alpha:
    ctr, _ = evaluate(test, verbose=2, mode="deploy")
    ctrs.append(ctr / rnd_ctr)

plt.plot(epsilon_values, ctrs)
plt.legend(["ε-greedy"])
plt.xlabel("ε")
plt.ylabel("CTR lift")
plt.title("Deployment bucket")
plt.show()
```

```python
alpha_values = np.arange(0.1,1.4,0.2)
tune_alpha_ucb = [ucb1(a) for a in alpha_values]

ctrs = []
for test in tune_alpha_ucb:
    ctr, _ = evaluate(test,verbose=3, mode="learn")
    ctrs.append(ctr / rnd_ctr)

plt.plot(alpha_values, ctrs)


tune_alpha_linucb = [linucb(a) for a in alpha_values]
ctrs = []
for test in tune_alpha_linucb:
    ctr, _ = evaluate(test,verbose=3, mode="learn")
    ctrs.append(ctr / rnd_ctr)

plt.plot(alpha_values, ctrs)
plt.legend(["ucb", "linucb"])
plt.xlabel("α")
plt.ylabel("CTR lift")
plt.title("Learning bucket")
plt.show()
```

```python
ctrs_ucb = []
for test in tune_alpha_ucb:
    ctr, _ = evaluate(test,verbose=2, mode="deploy")
    ctrs_ucb.append(ctr / rnd_ctr)
plt.plot(alpha_values, ctrs_ucb)

ctrs_linucb = []
for test in tune_alpha_linucb:
    ctr, _ = evaluate(test,verbose=2, mode="deploy")
    ctrs_linucb.append(ctr / rnd_ctr)
plt.plot(alpha_values, ctrs_linucb)

plt.legend(["ucb", "linucb"])
plt.xlabel("α")
plt.ylabel("CTR lift")
plt.title("Deployment bucket")
plt.show()
```
